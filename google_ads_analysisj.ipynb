```python
import pandas as pd

# Importar engine de la libreria sqlalchemy
from pandasql import sqldf
# Funcion lambda para correr querrys mas facil
pysqlc = lambda q: sqldf(q, globals())
```

Importar los datos

```python
df = pd.read_csv("google_ads_data.csv")
```

```python
df["Keyword"].duplicated().sum()
```

Eliminar todas las filas que contengan valores nulos con la funcion dropna

```python
df = df.dropna(subset=["Competition (indexed value)"])
```

# SQL y exportacion

```python
df = df.rename(columns={"Avg. monthly searches": "avg_searches", "Competition (indexed value)": "comp_value"})
```

```python
dtb= """
SELECT
Keyword,
avg_searches,
comp_value
FROM ar
where avg_searches >= 500
"""

df_export = pysqlc(dtb)
```

```python
df_export.to_csv("google_ads_data_processed.csv")
```
